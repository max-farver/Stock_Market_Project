```python
import alpha_vantage.timeseries as av
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

ts = av.TimeSeries(key='1W79PCAA71QM8Q85', output_format='pandas')
```

```python
# Example from alpha_vantage docs

data, meta_data = ts.get_intraday(symbol='AMD',interval='1min', outputsize='full')
data['4. close'].plot()
plt.title('Intraday Times Series for the MSFT stock (1 min)')
plt.locator_params(nbins=4)
plt.xticks(rotation=90)
plt.show()
```

```python
data, meta_data = ts.get_daily(symbol='AMD')
data['4. close'].plot()
plt.title('Intraday Times Series for the MSFT stock (1 min)')
plt.locator_params(nbins=4)
plt.xticks(rotation=90)
plt.show()
```
